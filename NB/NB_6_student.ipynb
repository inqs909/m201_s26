## Introduction

The `ca_housing` data set contains census block-level data  about housing information in California in 1990. The data set contains the following variables:

- `longitude`
- `latitude`
- `housing_median_age`: Median average age of home in the census block.
- `total_rooms`: Total number of rooms in the census block.
- `total_bedrooms`: Total number of bedrooms in the census block.
- `population`: the size of of population in the census block.
- `households`: The number of households available in the census block.
- `median_income`: The median household income in the census block.
- `median_house_value`: The median house price in the census block.
- `ocean_proximity`: How close is the house from the ocean?
