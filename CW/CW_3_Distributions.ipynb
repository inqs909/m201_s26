## Part 1: Random Number Generation

Below, we will simulate the following distributions and create a density plot to visualize what we see.

$$
X \sim N(8, 2)
$$

```r
# rnorm(1, 8, 2)
x <- rnorm(1000, 8, 2)
ggplot(data.frame(x), aes(x)) +
  geom_density() +
  stat_function(fun = dnorm,
                args = list(mean = 8, sd = 2),
                col = "red")
```

$$
X \sim \chi^2_4
$$

```r
# rchisq(1, 4)
x <- rchisq(1000, 4)
ggplot(data.frame(x), aes(x)) +
  geom_density()+
  stat_function(fun = dchisq,
                args = list(df = 4),
                col = "red")
```

$$
X \sim Bin(20, 0.35)
$$

```r
nn <- 20
pp <- 0.35
x <- rbinom(1000, size = nn, prob = pp)
df <- data.frame(x = x)
k <- 0:nn 
binom_df <- data.frame(
  x = k,
  prob = dbinom(k, size = nn, prob = pp)
)

ggplot(df, aes(x)) +
  geom_histogram(aes(y = after_stat(density)),
                 binwidth = 1,
                 alpha = 0.5) +
  geom_step(data = binom_df,
            aes(x = x - 0.5, y = prob),
            color = "red") 
```

$$
X \sim Pois(6)
$$

```r
# rpois(100, 6)
x <- rpois(1000, 6)
df <- data.frame(x = x)
k <- 0:max(x) 
pois_df <- data.frame(
  x = k,
  prob = dpois(k, 6)
)

ggplot(df, aes(x)) +
  geom_histogram(aes(y = after_stat(density)),
                 binwidth = 1,
                 alpha = 0.5) +
  geom_step(data = pois_df,
            aes(x = x - 0.5, y = prob),
            color = "red") 
```
