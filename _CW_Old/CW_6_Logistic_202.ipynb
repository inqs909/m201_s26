## Problem 1: Heart Transplant

We are interested in learning about the survival rate among patients who recieved a heart transplant. The data (`heart_transplant`) is based on the Stanford University Heart Transplant study. More information about the data can be found [here](https://www.openintro.org/data/index.php?data=heart_transplant). We will primarily focusing on explaining our outcome `survived`.

```r
glimpse(heart_transplant)
```

```r
cat_stats(heart_transplant$survived)
```

```r
num_stats(heart_transplant$age)
```

```r
glm(survived ~ age, data = heart_transplant, family = binomial)
```

$$
\log(odds\ of\ death) = \beta_0 + \beta_1 (age)
$$
$$
lo(death) = -1.56 + 0.06 (age)
$$


```r
exp(0.06)
```

As age increase by 1 year, the odds of dying increases by a factor of 1.06.

```r
cat_stats(heart_transplant$transplant)
```

```r
glm(survived ~ transplant,
data = heart_transplant,
family = binomial)
```

$$
lo(dying) = 2.015 - 1.39 (treatment)
$$

```r
exp(-1.39)
```

The odds of dying is 0.248 times lower for a patient who has had a transplant than a patient who has not (control group).

```r
cat_stats(heart_transplant$prior)
```

```r
glm(survived ~ prior,
data = heart_transplant,
family = binomial)
```

$$
lo(dying) = 1.143 - 1.143 (yes)
$$

```r
exp(-1.143)
```

For a patient who has had prior surgery, their odds of dying is 0.31 times lower than a patient who has not had surgery.

```r
glm(survived ~ transplant + age + prior,
data = heart_transplant,
family = binomial)
```

$$
lo(dying) = -1.02 - 1.67 (transplant) + 0.08 (age) - 0.84 (yes)
$$

```r
exp(-1.67)
```

What is the probability of dying for a patient who has had a transplant, is 45 years, and has had no prior surgery?

```r
m <- glm(survived ~ transplant + age + prior, data = heart_transplant, family = binomial())
# Predict probabilities
new_df <- data.frame(transplant = "treatment", age = 45, prior = "no")
predict(m, newdata = new_df, type = "response")
```

```r
m <- glm(survived ~ transplant + age + prior, data = heart_transplant, family = binomial())
# Predict probabilities
new_df <- data.frame(transplant = "control", age = 45, prior = "no")
predict(m, newdata = new_df, type = "response")
```

The odds of dying in a patient who has had a transplant is 0.18 times lower than a patient who has not had a transplant, adjusting for age and prior surgery.
