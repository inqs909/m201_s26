## Problem 1: California Housing

The `ca_housing` data set contains block-level data about housing information. We are interested in evaluating `median_house_value` (outcome of interest) by `housing_median_age`, `median_income` (median household income), `total_rooms`, and `ocean_proximity`.



```r
lm(median_house_value ~ median_income, data = ca_housing)
```

```r
cat_stats(ca_housing$ocean_proximity)
```

```r
lm(median_house_value ~ ocean_proximity, data = ca_housing)
```

```r
lm(median_house_value ~ total_rooms, data = ca_housing)
```

```r
lm(median_house_value ~ housing_median_age, data = ca_housing)
```

```r
lm(median_house_value ~ median_income + housing_median_age + total_rooms + ocean_proximity,
   data = ca_housing)
```

$$
value = \beta_0 + \beta_1 (income) + \beta_2 (age) + \beta_3 (rooms) + \beta_4 (inland) + \beta_5 (island) + \beta_6 (bay) + \beta_7 (ocean)
$$

What is the value of home that median income is 7.5, median age 70, rooms is 120, and is inland?

$$
value = 38570 + 37562 (7.5) + 1143 (70) + 3.467 (120) - 71484 (1) + 184832 (0) + 12101 (0) + 17874 (0)
$$

```r
38570 + 37562 * 7.5 + 1143 * 70 + 3.467 * 120 - 71484 * 1
```

```r
m <- lm(median_house_value ~ median_income + housing_median_age + total_rooms + ocean_proximity,
   data = ca_housing)
predict(m, newdata = data.frame(median_income = 7.5,
            housing_median_age = 70, total_rooms = 120,
            ocean_proximity = "INLAND"))
```

$$
value = 38570 + 37562 (income) + 1143 (age) + 3.467 (rooms) - 71484 (inland) + 184832 (island) + 12101 (bay) + 17874 (ocean)
$$

As the income increases by 1 unit, the average household value increases by an average 37562, adjusting for rooms, age, and ocean proximity.

On average, a household inland is 71484 dollars lower than a househod that is <1 hour from the ocean, adjusting for income, age, and rooms.

On average, a household located near the ocean is 17874 dollars higher than a household that is < 1 hour from the ocean, adjusting for income, age, and rooms.

```r
lm(median_house_value ~ median_income + housing_median_age + total_rooms + ocean_proximity,
   data = ca_housing)
lm(median_house_value ~ ocean_proximity + median_income + housing_median_age + total_rooms ,
   data = ca_housing)
```
