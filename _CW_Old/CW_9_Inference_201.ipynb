## Problem 1: California Housing

The `ca_housing` data set contains block-level data about housing information. We are interested in evaluating if there is a significant association between `median_house_value` (outcome of interest) and `housing_median_age` or `median_income` (median household income).



```r
glimpse(ca_housing)
```

$$
med\_value = \beta_0 + \beta_1 (med\_age)
$$

```r
lm(median_house_value ~ housing_median_age, data = ca_housing)
```

$$
med\_value = 179119 + 968.4 (med\_age)
$$

Is there a significant relationship between median value and median age?

$$
H_0: \beta_1 = 0
$$
$$
H_1: \beta_1 \ne 0
$$


$$
\alpha = 0.05
$$

```r
m <- lm(median_house_value ~ housing_median_age, data = ca_housing)
broom::tidy(m)
```

$$
p = 2.76 e-52 = 2.76 * 10 ^{-52}
$$

$$
\alpha = 0.05 = 5.0*10^{-2}
$$

$$
p < \alpha
$$

Reject the Null Hypothesis

There is a significant relationship between median house value and median house age, (p < 0.05). As median house age increase by 1 unit, the median house value increases by 968 units.
