## Problem 1: California Housing

The `ca_housing` data set contains block-level data about housing information. We are interested in evaluating `median_house_value` (outcome of interest) by `housing_median_age`, `median_income` (median household income), `total_rooms`, and `ocean_proximity`. Is there a significant association between `housing_median_age` and `median_house_value`, adjusting for `total_rooms` and `ocean_proximity`?



```r
glimpse(ca_housing)
```

```r
m1 <- lm(median_house_value ~ housing_median_age + total_rooms + median_income + ocean_proximity,
         data = ca_housing)
```

```r
m1
```

$$
housing\_price = 38570 + 1143 (age) + 3.467 (rooms) + 37562 (income) - 71484 (inland) + 184832 (island) +12101 (nearbay) + 17874 (near ocean)
$$

$H_0$: There is no association between median_house_age and median_house_price, adjusting rooms, income, and ocean proximity ($\beta_{age} = 0$)

$H_1$: There is an association between median_house_age and median_house_price, adjusting rooms, income, and ocean proximity ($\beta_{age} \ne 0$)

$$
\alpha = 0.05 = 5e-2=5*10^{-2}
$$

```r
tidy(m1)
```

$$
p = 5.32e-134 = 5.32*10^{-134}
$$

$$
p < \alpha
$$


## Reject $H_0$

There is a significant association betweem median house value and median house age, adjusting for total rooms, median income and ocean proximity (p < 0.0001). As median house age increases by 1 unit, the median house value increases by 1143 units, adjusting for total rooms, median income, and ocean proximity.
