### Central Tendency

```r
# Mean
mean(Running$HRM)
# Median
median(Running$HRM)
```

### Variation/Variability

```r
# Standard Deviation
sd(Running$HRM)
# Range
min(Running$HRM)
max(Running$HRM)
# Variance
var(Running$HRM)
sd(Running$HRM)^2
# Interquartile Range
quantile(Running$HRM, probs = c(0.25, 0.75))
```

```r
num_stats(Running$HRM)
```

The average VO for an HRM wearable is 108.389 with a standard deviation of 18.504. The median VO for an HRM wearable is 112.09 with an interquarile range of 27.575 (Q1: 94.132 and Q3: 121.708). With a minimum and maximum of 63.71 and 140.67, respectively.
