## CLT 1

Using the formulas below, we will simulate multiple samples and compute the mean of each sample. Afterward, we will plot the density plot of the means.

$$
X \sim N(8, 2)
$$

$$
E(X) = 8\ \ SD(X) = 2
$$

```r
x <- replicate(10000, rnorm(1000, 8, 2))
x_means <- colMeans(x)
ggplot(data.frame(x = x_means), aes(x)) +
  geom_density(col = "blue") +
  stat_function(fun = dnorm,
                args = list(mean = 8, sd = 2 / sqrt(1000)),
                col = "red")
```
