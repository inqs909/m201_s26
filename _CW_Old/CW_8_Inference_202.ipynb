## Problem 1: California Housing

The `ca_housing` data set contains block-level data about housing information. We are interested in evaluating if there is a significant association between `median_house_value` (outcome of interest) and `housing_median_age` or `median_income` (median household income).



$$
med\_value = \beta_0 + \beta_1 (med\_age)
$$

```r
lm(median_house_value ~ housing_median_age, data = ca_housing)
```

$$
med\_value = 179119.9 + 968.4(med\_age)
$$

$H_0$: There is no association between median house value and housing median age ($\beta_1 = 0$)

$H_1$: There is an associationa between median house value and housing median age ($\beta_1 \ne 0$)

$$
\alpha = 0.05
$$

```r
m1 <- lm(median_house_value ~ housing_median_age, data = ca_housing)
tidy(m1)
```

$$
p = 2.76e-52 = 2.76*10^{-52}
$$
$$
\alpha = 5.0 *10^{-2}
$$
$$
p < \alpha
$$


## Reject $H_0$ (Null is incorrect)

There is a significant relationship between median house value and median house age (p < 0.001). As median house age increases by 1 year, the median house value increases by 968 units.

## Imagine: Fail to Reject

There is no significant relationship between median house value and median house age (p > 0.05).

## Problem 2: A Dogs Life

The `DogsLife` data set contains information about different dog breeds lifespan. You can find more information about the data set [here](https://cran.r-project.org/web/packages/SRMData/refman/SRMData.html#DogsLife) and their reference paper [here](https://www.journals.uchicago.edu/doi/10.1086/724384) We are interested if there is a significant association between `Lifespan` (outcome) and  `Weight` or `LitterSize`.



$$
Lifespan = \beta_0 + \beta_1 (Weight)
$$

```r
lm(Lifespan ~ Weight, data = DogsLife)
```

$$
Lifespan = 13.38 - 0.09(Weight)
$$

$H_0$: There is no relationship between Dog Weight and Lifespan ($\beta_1 = 0$).

$H_1$: There i a relationship between Dog Weight and Lifespan ($\beta_1 \ne 0$)

$$
\alpha = 0.05
$$


```r
m2 <- lm(Lifespan ~ Weight, data = DogsLife)
tidy(m2)
```

$$
p = 1.68e-12  <  5e-2 = \alpha
$$

## Reject the $H_0$ (Null Incorrect)

There is a significant association between lifespan and weight (p < 0.05). As weight increases by 1 unit, lifespan decreases by an average of 0.09 units.
