## Problem 1

The `bdims` data set contains information of body and skeletal measurements.

### Part A

Run the code below to read the help documentation or click [here](https://www.openintro.org/book/statdata/?data=bdims). How many observations and variables are in the data set?

```r
?bdims
```

### Part B

We are interested in analyzing the relationship between the outcome variable `elb_di` (elbow diameter) and predictor `hgt` (height). Describe the distribution of both variables.

```r
ggplot(bdims, aes(elb_di)) +
  geom_density()
```

### Part C

Use `hgt` and `elb_di` and create a scatter plot. Describe the association between the variables.

```r
ggplot(bdims, aes(x = hgt, y = elb_di)) +
  geom_point()
```

### Part D

Add a trendline to the scatter plot above.

```r
ggplot(bdims, aes(x = hgt, y = elb_di)) +
  geom_point() +
  stat_smooth(method = "lm")
```

### Part E

Fit a regression line between the outcome variable `elb_di` and `hgt`. Write out the equation of the model.

$$
elb\_di = \beta_0 + \beta_1 (hgt)
$$

```r
lm(elb_di ~ hgt, data = bdims)
```

### Part F

Interpret the Intercept and Slope.

When there is a height of 0 cm, the average elbow diameter is - 4.7821 cm.

As the height increases by 1 cm, the average elbow diameter increases by 0.1062 cm.

$$
elb\_di = -4.7821 + 0.1062 (hgt)
$$

### Part G

What is the average elbow diameter when a person is 60 centimeters?

```r
-4.7821 + 0.1062*60
```

```r
m <- lm(elb_di ~ hgt, data = bdims)
predict(m, newdata = data.frame(hgt = 60))
```
